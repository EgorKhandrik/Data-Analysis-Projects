## Исследовательский анализ данных (EDA)

В рамках исследовательского анализа данных изучим в целом представленные данные и информацию о поведении пользователей. **Будем ориентироваться в этом блоке, прежде всего, на исследование календарных недель, без привязки к когортам.**

### Создаём профили

Напишем функцию `get_profiles` для создания профилей юзеров:

```python
def get_profiles(sessions):

    # находим параметры первых посещений
    profiles = (
        sessions.sort_values(by=['user_id', 'event_time'])
        .groupby('user_id')
        .agg(
            {
                'event_time': 'first',
                'source': 'first',
            }
        )
        .rename(columns={'event_time': 'first_ts'})
        .reset_index()
    )

    # для когортного анализа определяем дату первого посещения,
    # первый день месяца и номер недели, в которые это посещение произошло
    profiles['dt'] = profiles['first_ts'].dt.date
    profiles['month'] = profiles['first_ts'].astype('datetime64[M]')
    profiles['week_cohort'] = profiles['first_ts'].dt.isocalendar().week

    # считаем количество уникальных пользователей
    # с одинаковыми источником и датой привлечения
    new_users = (
        profiles.groupby(['dt', 'source'])
        .agg({'user_id': 'nunique'})
        .rename(columns={'user_id': 'unique_users'})
        .reset_index()
    )

    return profiles
```

Применим функцию `get_profiles` к нашим данным и сохраним результат в переменной `profiles`:

```python
profiles = get_profiles(data)
profiles.head()
```

Добавим в датафрейм `data` столбцы из `profiles`:

```python
data = data.merge(profiles, on='user_id', how='left')
data.head()
```

### Сколько сессий в среднем юзер проводит в день, неделю?

Определим сколько сессий юзер в среднем проводит в приложении в день, неделю.    
Для начала создадим столбец *time_diff*, в котором сохраним информацию о времени (в секундах), прошедших между соседними событиями юзера:

```python
# в столбце time_diff сохраним информацию о времени, которое прошло между последовательными событиями юзера
data['time_diff'] = (
                     data
                     .sort_values(by=['user_id', 'event_time'])
                     .groupby('user_id')['event_time']
                     .diff()
                     .dt.total_seconds()
)
```

Будем считать за сессию если между событиями юзера в приложении прошло не более 10 минут. В столбце *session* будем сохранять данные о том в какой по порядку сессии совершены действия юзера: 

```python
data['session'] = (data
                   .sort_values(by=['user_id', 'event_time'])
                   .groupby('user_id')['event_time']
                   .transform(lambda x: (x.diff() >= '00:10:00')
                   .cumsum())
                  )

data.head()
```

Также создадим столбцы *date* и *week* с информацией о дате и неделе, в которую совершалось действие пользователя:

### Сколько в среднем длятся сессии

Узнаем сколько длятся сессии юзеров в приложении. В перемнной `sessions` сохраним датафрейм с информацией о всех сессиях юзеров. В столбце *duration_sec* сохраним сведения о длительности сессии в секундах:

```python
sessions = (
    data
    .pivot_table(
        index=['user_id', 'session'],
        values='event_time',
        aggfunc=['first', 'last'])
    .reset_index()
    .droplevel(1, axis=1)
)

sessions.columns = ['user_id', 'session', 'session_start', 'session_end']

sessions['duration_sec'] = (sessions['session_end'] - sessions['session_start']).dt.total_seconds()

display(sessions['duration_sec'].describe())
```

Имеется очень много данных о сессиях длительностью в ноль секунд. Предположим, что в эти сессии пользователем совершалось одно действие и затем приложение закрывалось.     
Выведем соотношение сессий с нулевой продолжительностью к более длительным сессиям. Для этого создадим столбец *one_action*, в котором **1** будет означать сессию с нулевой продолжительностью, а **0** - сессию с более чем нуевой длительностью.

```python
sessions['one_action'] = np.where(sessions['duration_sec'] == 0, 1, 0)
sessions['one_action'] = sessions['one_action'].astype(int)

sessions.groupby('one_action').agg('count').plot(y='user_id', kind='pie', autopct='%.0f%%')
plt.title('Соотношение сессий', fontsize=17)
plt.legend(bbox_to_anchor = (1.0, 1), loc = 'best')
plt.ylabel('');
```

Таким образом, 26% от всех сессий длятся ноль секунд. Узнаем какие события чаще всего попадают в нулевые по длительности сессии.

```python
zero = sessions.query('one_action == 1')
zero = data.merge(zero[['user_id', 'session', 'duration_sec', 'one_action']], on=['user_id', 'session'])
zero['event_name'].value_counts()
```

Больше всего событий в сессиях с нулевой длительностью приходится на показ рекомендованных объявлений и просмотр фото, что не отличается от данных в целом. Взаимосвязь нулевых сессий с конкретными событиями не установлена.

### Распределение количества событий по дням недели и времени суток

Оценим как распределяется количество действий юзеров в приложении по времени суток. Для этого добавим столбец `hour` с информацией о часе, в котором совершено действие. Далее исходя из часа совершения действия определим в какое времени суток было совершено действие. Данные о времени суток сохраним в колонке `period`. Временами суток будем считать:
* ночь (night) - с 23:00:00 до 05:59:59;
* утро (morning) - с 6:00:00 до 11:59:59;
* день (afternoon) - с 12:00:00 до 17:59:59;
* вечер (evening) - с 18:00:00 до 22:59:59.

```python
data['hour'] = data['event_time'].dt.hour

condition = [
    (data['event_time'].dt.hour >= 23) | (data['event_time'].dt.hour < 6),
    (data['event_time'].dt.hour >= 6) & (data['event_time'].dt.hour < 12),
    (data['event_time'].dt.hour >= 12) & (data['event_time'].dt.hour < 18),
    (data['event_time'].dt.hour >= 18) & (data['event_time'].dt.hour < 23)
]

values = ['night', 'morning', 'afternoon', 'evening']

data['period'] = np.select(condition, values)

data.groupby('period').agg('count').plot(y='user_id', kind='pie', autopct='%.0f%%')
plt.title('Количество событий по времени суток', fontsize=17)
plt.legend(bbox_to_anchor = (1.0, 1), loc = 'best')
plt.ylabel('');
```

Более 70% событий совершаются юзерами днём (41%) и вечером (31%). Утром происходит 19% событий в приложении, а ночь приходится всего 9%.

### Сколько сессий заканчивается просмотром контактов

Добавим данные о совершении в течение сессии целевого события - просмотра контактов в объявлении. Сохраним данные в столбец `target_action`, в которой **1** - целевое действие в течение сессии было совершено, **0** - целевое действие не совершалось в рамках сессии.

```python
target = (
    data
    .groupby(['user_id', 'session'])['event_name']
    .agg('unique')
)

target = target.to_frame().reset_index()
lst = ['contacts_show']
target['target_action']  = np.where(target['event_name'].apply(lambda s: len(set(s) & set(lst)) > 0), 1, 0)
target.head(10)
```

Добавим колонку `target_action` в датафрейм `sessions`:

```python
rt = rt.groupby('user_id').agg({'lifetime':['max', 'nunique']})
rt = rt.reset_index().droplevel(1, axis=1)
rt.columns = ['user_id', 'max_lifetime', 'lifetime_days']
profiles = profiles.merge(rt[['user_id', 'max_lifetime', 'lifetime_days']], on='user_id', how='left')
```

Визуализируем удержание также с помощью кривых. Для начала создадим необходимые функции:

```python
# функция для сглаживания фрейма

def filter_data(df, window):
    # для каждого столбца применяем скользящее среднее
    for column in df.columns.values:
        df[column] = df[column].rolling(window).mean() 
    return df 
```

```python
def get_retention(
    profiles,
    sessions,
    observation_date,
    horizon_days,
    dimensions=[],
    ignore_horizon=False,
):

    # добавляем target_action в передаваемый dimensions список
    dimensions = ['target_action'] + dimensions

    # исключаем пользователей, не «доживших» до горизонта анализа
    last_suitable_acquisition_date = observation_date
    if not ignore_horizon:
        last_suitable_acquisition_date = observation_date - timedelta(
            days=horizon_days - 1
        )
    result_raw = profiles.query('dt <= @last_suitable_acquisition_date')
    # собираем «сырые» данные для расчёта удержания
    result_raw = result_raw.merge(
        sessions[['user_id', 'session_start', 'target_action']], on='user_id', how='left'
    )
    result_raw['lifetime'] = (
        result_raw['session_start'] - result_raw['first_ts']
    ).dt.days

    # функция для группировки таблицы по желаемым признакам
    def group_by_dimensions(df, dims, horizon_days):
        result = df.pivot_table(
            index=dims, columns='lifetime', values='user_id', aggfunc='nunique'
        )
        cohort_sizes = (
            df.groupby(dims)
            .agg({'user_id': 'nunique'})
            .rename(columns={'user_id': 'cohort_size'})
        )
        result = cohort_sizes.merge(result, on=dims, how='left').fillna(0)
        result = result.div(result['cohort_size'], axis=0)
        result = result[['cohort_size'] + list(range(horizon_days))]
        result['cohort_size'] = cohort_sizes
        return result
    # получаем таблицу удержания
    result_grouped = group_by_dimensions(result_raw, dimensions, horizon_days)

    # получаем таблицу динамики удержания
    result_in_time = group_by_dimensions(
    result_raw, dimensions + ['dt'], horizon_days
    )

    # возвращаем обе таблицы и сырые данные
    return result_raw, result_grouped, result_in_time 
```

```python
# функция для визуализации удержания

def plot_retention(retention, retention_history, horizon, window=7):

    # задаём размер сетки для графиков
    plt.figure(figsize=(15, 10))

    # исключаем размеры когорт и удержание первого дня
    retention = retention.drop(columns=['cohort_size', 0])
    # в таблице динамики оставляем только нужный лайфтайм
    retention_history = retention_history.drop(columns=['cohort_size'])[
        [horizon - 1]
    ]

    # если в индексах таблицы удержания только target_action,
    # добавляем второй признак — cohort
    if retention.index.nlevels == 1:
        retention['cohort'] = 'All users'
        retention = retention.reset_index().set_index(['cohort', 'target_action'])
        # в таблице графиков — два столбца и две строки, четыре ячейки
    # в первой строим кривые удержания "сконвертировавшихся" пользователей
    ax1 = plt.subplot(2, 2, 1)
    retention.query('target_action == 1').droplevel('target_action').T.plot(
        grid=True, ax=ax1
    )
    plt.legend()
    plt.xlabel('Лайфтайм')
    plt.title('Удержание пользователей, \n совершивших целевое действие (просмотр контактов)')

    # во второй ячейке строим кривые удержания "несконвертировавшихся"
    # вертикальная ось — от графика из первой ячейки
    ax2 = plt.subplot(2, 2, 2, sharey=ax1)
    retention.query('target_action == 0').droplevel('target_action').T.plot(
        grid=True, ax=ax2
    )
    plt.legend()
    plt.xlabel('Лайфтайм')
    plt.title('Удержание пользователей, \n не совершивших целевое действие (просмотр контактов)')

    # в третьей ячейке — динамика удержания "сконвертировавшихся"
    ax3 = plt.subplot(2, 2, 3)
    # получаем названия столбцов для сводной таблицы
    columns = [
        name
        for name in retention_history.index.names
        if name not in ['dt', 'target_action']
    ]
    # фильтруем данные и строим график
    filtered_data = retention_history.query('target_action == 1').pivot_table(
        index='dt', columns=columns, values=horizon - 1, aggfunc='mean'
    )
    filter_data(filtered_data, window).plot(grid=True, ax=ax3)
    plt.xlabel('Дата привлечения')
    plt.title(
        'Динамика удержания пользователей, \n совершивших целевое действие (просмотр контактов) \n на {}-й день'.format(
            horizon
        )
    )

    # в чётвертой ячейке — динамика удержания "несконвертировавшихся"
    ax4 = plt.subplot(2, 2, 4, sharey=ax3)
    # фильтруем данные и строим график
    filtered_data = retention_history.query('target_action == 0').pivot_table(
        index='dt', columns=columns, values=horizon - 1, aggfunc='mean'
    )
    filter_data(filtered_data, window).plot(grid=True, ax=ax4)
    plt.xlabel('Дата привлечения')
    plt.title(
        'Динамика удержания \n не совершивших целевое действие (просмотр контактов) на {}-й день'.format(
            horizon
        )
    )
    plt.tight_layout()
    plt.show() 
```

Визуализируем кривые удержания по недельным когортам. Определим моментом анализа 4 ноября 2019 г., а горизонтом анализа 7 дней. В качестве целевого действия будем считать просмотр контактов юзером (*contacts_show*).

Посмотрим как изменяется продолжительость сессий по временам суток. Временами суток будем считать:
* ночь (night) - с 23:00:00 до 05:59:59;
* утро (morning) - с 6:00:00 до 11:59:59;
* день (afternoon) - с 12:00:00 до 17:59:59;
* вечер (evening) - с 18:00:00 до 22:59:59.

```python
condition = [
    (sessions['session_start'].dt.hour >= 23) | (sessions['session_start'].dt.hour < 6),
    (sessions['session_start'].dt.hour >= 6) & (sessions['session_start'].dt.hour < 12),
    (sessions['session_start'].dt.hour >= 12) & (sessions['session_start'].dt.hour < 18),
    (sessions['session_start'].dt.hour >= 18) & (sessions['session_start'].dt.hour < 23)
]

values = ['night', 'morning', 'afternoon', 'evening']

sessions['period'] = np.select(condition, values)
```

### Conversion Rate (в просмотр контактов)

Создадим функции для визуализации кривых конверсии:

```python
# функция для расчёта конверсии

def get_conversion(
    profiles,
    purchases,
    observation_date,
    horizon_days,
    dimensions=[],
    ignore_horizon=False,
):

    # исключаем пользователей, не «доживших» до горизонта анализа
    last_suitable_acquisition_date = observation_date
    if not ignore_horizon:
        last_suitable_acquisition_date = observation_date - timedelta(
            days=horizon_days - 1
        )
    result_raw = profiles.query('dt <= @last_suitable_acquisition_date')

    # группируем по cohort, если в dimensions ничего нет
    if len(dimensions) == 0:
        result_raw['cohort'] = 'All users' 
        dimensions = dimensions + ['cohort']

    # функция для группировки таблицы по желаемым признакам
    def group_by_dimensions(df, dims, horizon_days):
        result = df.pivot_table(
            index=dims, columns='lifetime', values='user_id', aggfunc='nunique'
        )
        result = result.fillna(0).cumsum(axis = 1)
        cohort_sizes = (
            df.groupby(dims)
            .agg({'user_id': 'nunique'})
            .rename(columns={'user_id': 'cohort_size'})
        )
        result = cohort_sizes.merge(result, on=dims, how='left').fillna(0)
        # делим каждую «ячейку» в строке на размер когорты
        # и получаем conversion rate
        result = result.div(result['cohort_size'], axis=0)
        result = result[['cohort_size'] + list(range(horizon_days))]
        result['cohort_size'] = cohort_sizes
        return result

    # получаем таблицу конверсии
    result_grouped = group_by_dimensions(result_raw, dimensions, horizon_days)

    # для таблицы динамики конверсии убираем 'cohort' из dimensions
    if 'cohort' in dimensions: 
        dimensions = []

    # получаем таблицу динамики конверсии
    result_in_time = group_by_dimensions(
        result_raw, dimensions + ['dt'], horizon_days
    )

    # возвращаем обе таблицы и сырые данные
    return result_raw, result_grouped, result_in_time 
```

```python
def plot_conversion(conversion, conversion_history, horizon, window=1):

    # задаём размер сетки для графиков
    plt.figure(figsize=(15, 5))

    # исключаем размеры когорт
    conversion = conversion.drop(columns=['cohort_size'])
    # в таблице динамики оставляем только нужный лайфтайм
    conversion_history = conversion_history.drop(columns=['cohort_size'])[
        [horizon - 1]
    ]

    # первый график — кривые конверсии
    ax1 = plt.subplot(1, 2, 1)
    conversion.T.plot(grid=True, ax=ax1)
    plt.legend()
    plt.xlabel('Лайфтайм')
    plt.title('Конверсия пользователей')

    # второй график — динамика конверсии
    ax2 = plt.subplot(1, 2, 2, sharey=ax1)
    columns = [
        # столбцами сводной таблицы станут все столбцы индекса, кроме даты
        name for name in conversion_history.index.names if name not in ['dt']
    ]
    filtered_data = conversion_history.pivot_table(
        index='dt', columns=columns, values=horizon - 1, aggfunc='mean'
    )
    filter_data(filtered_data, window).plot(grid=True, ax=ax2)
    plt.xlabel('Дата привлечения')
    plt.title('Динамика конверсии пользователей на {}-й день'.format(horizon))

    plt.tight_layout()
    plt.show() 
```

```python
dimensions = ['source']

conversion_raw, conversion_grouped, conversion_history = get_conversion(
    profiles, sessions, observation_date, horizon_days, dimensions=dimensions
)

plot_conversion(conversion_grouped, conversion_history, horizon_days) 
```

* Юзеры, привлечённые через Яндекс и Google показывают лучшую конверсию в целевое действие, чем юзеры из иных источников;
* Пользователи из Яндекса, привлеченные в первые 10 дней исследуемого периода, показывали лучшую конверсию через 7 дней с момента начала пользования приложением. Однако к концу периода юзеры из Google продемонстрировали большую конверсию. 

Построим также кривые в разрезе когорт:

```python
dimensions = ['week_cohort']

conversion_raw, conversion_grouped, conversion_history = get_conversion(
    profiles, sessions, observation_date, horizon_days, dimensions=dimensions
)

plot_conversion(conversion_grouped, conversion_history, horizon_days) 
```

## Сегментация

Произведём сегментацию юзеров с помощью метода k-means. Выделим кластеры:
* по времени, проведённому в приложении;
* по действиям, совершённым за сессию;
* по удержанию (частоте использования приложения).
      
Создадим функцию для визуализации elbow method и silhouette method, которые помогут определить оптимальное количество кластеров в каждом планируемом сегменте, а также функцию для осуществления кластеризации:

```python
sc = StandardScaler()

# функция для elbow method
def elbow(cols, title):
    clusters = profiles[cols]
    clusters_sc = sc.fit_transform(clusters)
    
    inertia = []
    for n_clusters in range(2, 15):
        kmeans = KMeans(n_clusters=n_clusters, random_state=0).fit(clusters_sc)
        inertia.append(kmeans.inertia_ / n_clusters)
    
    inertias = pd.DataFrame({'n_clusters': range(2, 15), 'inertia': inertia})
    ax = inertias.plot(x='n_clusters', y='inertia', figsize=[7, 5])
    plt.title(title)
    plt.xlabel('Количество кластеров')
    plt.ylabel('Инерция')
    plt.ylim((0, 1.1 * inertias.inertia.max()))
    ax.legend().set_visible(False)
    plt.tight_layout()
    plt.show();

def silhouette(cols, title):
    nK = 12 #количество кластеров
    
    clusters = profiles[cols]
    clusters_sc = sc.fit_transform(clusters)
    
    sil_scores = [silhouette_score(
                                clusters_sc, 
                                KMeans(n_clusters=k).
                                fit_predict(clusters_sc))
                                for k in tqdm(range(2,nK))]
    dict_sil = dict(zip(range(2,nK), sil_scores))
    print("silhouette scores:")
    _ = [print(k, ":", f'{v:,.3f}') for k,v in dict_sil.items()]
    K_sil_a3 = max(dict_sil, key=dict_sil.get) #оптимальное количество класетров
    sil_opt_a3 = dict_sil[K_sil_a3]            #оптимальная оценка силуета
    print("Максимальный silhouette score для", f'{K_sil_a3:.0f} кластеров: ', 
          f'{sil_opt_a3:.3f}')
    
    plt.figure(figsize=[7,5])
    plt.plot(range(2,nK), sil_scores, color="red")
    plt.title(title)
    plt.xticks(np.arange(2,nK,1))
    plt.xlabel("Количество кластеров")
    plt.ylabel("silhouette score")
    plt.show()
    
# функция для кластеризации
def clusters(k, cols, name):
    clusters = profiles[cols]
    clusters_sc = sc.fit_transform(clusters)
    clusters = KMeans(n_clusters = k).fit(clusters_sc)
    labels = clusters.labels_
    profiles[name] = clusters.labels_
    display(profiles.groupby(name)[cols].describe().T)
    return print(Counter(clusters.labels_))
```

В качестве признаков возьмём данные об общем количестве времени, проведённого в приложении (*sum_time_spent*), среднего времени сессии (*mean_time_per_session*) и медианного времени за сессию (*median_time_per_session*):

### Как различается время между распространенными событиями пользователей

Оценим отличаются ли выделенные сегменты по времени, которое они проводят от первого действия в приложении до поиска.    
Сохраним в переменной `search` датафрейм, который будет содержать информацию о времени первого действия пользователя в приложении и о времени первого действия по поиску. Также рассчитаем разницу в секундах, минутах и часах между первым действием и первым поиском.

```python
search = (
    data
    .merge(profiles[['user_id', 'rt_cluster', 'time_cluster', 'ta_cluster',
                     'action_cluster', 'target_action']], on='user_id')
)

search[['time_cluster', 'action_cluster', 'rt_cluster', 'ta_cluster']] = (
    search[['time_cluster', 'action_cluster', 'rt_cluster', 'ta_cluster']]
    .astype(str)
)

time_claster_replacer = {'0':'mean twenty min per session',
                         '1':'mean five min per session'}

for k, v in time_claster_replacer.items():
    search['time_cluster'] = search['time_cluster'].str.replace(k, v)

action_cluster_replacer = {'0':'mean twenty actions per session',
                           '1':'mean five actions per session'}

for k, v in action_cluster_replacer.items():
    search['action_cluster'] = search['action_cluster'].str.replace(k, v)

rt_cluster_replacer = {'0':'do not return very',
                       '1':'very often eleven and more days',
                       '2':'rarely three-five days',
                       '3':'often seven-ten days',
                       '4':'rarely two days'}

for k, v in rt_cluster_replacer.items():
    search['rt_cluster'] = search['rt_cluster'].str.replace(k, v)

ta_cluster_replacer = {'0':'do not click show_contacts',
                      '1':'every session with ta',
                      '2':'half of sessions with ta',
                      '3':'third of sessions with ta'}

for k, v in ta_cluster_replacer.items():
    search['ta_cluster'] = search['ta_cluster'].str.replace(k, v)

# переименуем все события по поиску к одному названию
search['event_name'] = (
    search['event_name']
    .replace(['search_1', 'search_2', 'search_3', 'search_4', 
              'search_5', 'search_6', 'search_7'], 'search')
)

search = search.query('event_name == "search"')

search = (
    search
    .groupby(['user_id', 'event_name', 'first_ts', 'rt_cluster', 
              'time_cluster', 'action_cluster', 'target_action'])
    .agg({'event_time':'first'})
    .reset_index()
)

search['diff_sec'] = (search['event_time'] - search['first_ts']).dt.total_seconds()
search['diff_min'] = (search['diff_sec'] / 60).astype(int)
search['diff_hours'] = (search['diff_min'] / 60).astype(int)
search.head()
```

Оценим получившиеся результаты:

```python
search[['diff_min']].value_counts(normalize=True).sort_values(ascending=False).head(10)
```

Как видим, очень много данных в которых первый поиск был в первую же минут (возможно, что и первым действием в приложении в принципе).     
Посмотрим поменяются ли значения, если мы ограничим данные первыми 24 часами, поскольку между первым действием и поиском может проходить несколько дней. 

```python
search_one_day = search.query('diff_hours <= 24')

display(
    search_one_day['diff_min']
    .value_counts(normalize=True)
    .sort_values(ascending=False)
    .head(10)
)

display('Количество данных с учётом времени до первого поиска 24+ часа: ', len(search))
display('Количество данных без учёта времени до первого поиска 24+ часа: ', len(search_one_day))
```

Учитывая, что данных с первым поиском более 24 часов не так много, не будем учитывать такие длинные периоды между началом использования приложения и первым поиском в дальнейшем.

Посмотрим отличаются ли выделенные сегменты по времени между началом пользования приложением и первым поиском.

```python
search_action = (
    search_one_day
    .groupby('action_cluster')['diff_min']
    .value_counts(normalize=True)
)

search_action = search_action.to_frame()
search_action.columns = ['min']
search_action = search_action.reset_index()
search_action.columns = ['action_cluster', 'min', 'percantage']

(
    search_action
    .sort_values(['action_cluster', 'percantage'], ascending=False)
    .groupby('action_cluster')
    .head(5)
)
```

* Юзеры, много совершающие действий за сессию, меньше доходят до посика в первые 60 секунд - всего 71% против 81% у сегмента мало совершающих действий за сессию. Предположим, что это связано с тем, что из-за большого количества действий за сессию использование поиска сдвигается по времени;
* Также возможно утверждать, что юзеры из первого сегмента по количеству действий в принципе позже доходят до поиска, т.к. в топ-5 попали значения в 6 и 9 минут между началом использования приложения и первым поиском, при этом у сегмента с меньшим количеством действий топ-5 ограничен 4 минутами.

Также оценим отличаются ли сегменты по времени, проведённому в приложении:

```python
search_time = (
    search_one_day
    .groupby('time_cluster')['diff_min']
    .value_counts(normalize=True)
)

search_time = search_time.to_frame()
search_time.columns = ['min']
search_time = search_time.reset_index()
search_time.columns = ['time_cluster', 'min', 'percantage']

(
    search_time
    .sort_values(['time_cluster', 'percantage'], ascending=False)
    .groupby('time_cluster')
    .head(5)
)
```

* У юзеров, проводящих больше времени в приложении, уходит больше времени с начала использования приложением до поиска: в первые 60 секунд поиск совершает 75% юзеров этого сегмента, при этом второе и третье значение - 6 и 9 минут соответственно. Кроме того, в топ-5 также попало 10 минут с начала использования приложения.
* Юзеры из сегмента, которые меньше проводят времени в приложении, быстрее доходят до поиска. В первые 60 секунд поиском пользуется 81% юзеров.

Дадим также оценку сегментам по удержанию:

## Проверка статистических гипотез

### Статистическая значимость различий в конверсии в целевое действие между юзерами, пришедшими из yandex и google

Проверим имеется ли статистическая значимость в конверсии в целевое действие между юзерами, пришедшими в приложение из yandex и google.

Для начала в переменной `cr_hyp_source` сохраним датафрейм с данными о количестве юзеров, впервые совершивших в каждый из анализируемых дней целевое действие:

```python
cr_hyp = cr.query('source == "yandex" | source == "google"')

cr_hyp_source = cr_hyp.pivot_table(
    index='source',
    columns='dt',
    values='first_ta',
    aggfunc='count'
)

cr_hyp_source
```

В переменной `sizes` сохраним данные о количестве юзеров из каждого источника на каждый из дней:

```python
sizes = (
    cr_hyp
    .groupby(['source', 'dt'])['user_id']
    .count()
    .to_frame()
    .reset_index()
)

sizes = (
    sizes
    .pivot_table(
        index='dt', 
        columns='source', 
        values='user_id')
    .reset_index()
    .set_index('dt')
    .T)

sizes
```

Рассчитаем конверсию на каждый из дней:

```python
cr_hyp_source = cr_hyp_source.div(sizes).T.reset_index()
cr_hyp_source.head()
```

Оценим распределения значений конверсий:

```python
fig, ax = plt.subplots()
ax.hist(cr_hyp_source['google'], alpha=0.5, label='google')
ax.hist(cr_hyp_source['yandex'], alpha=0.5, label='yandex')
ax.set(title='Распределение значений конверсий по источникам привлечения')
plt.legend();
```

Как видим, значения конверсий распределены ненормально.

В переменный `google` и `yandex` сохраним результаты конверсий для каждого источника:

```python
google = cr_hyp_source['google']
yandex = cr_hyp_source['yandex']
```

Оценку гипотез проведём с помощью тесте Манна-Уитни, поскольку:
* данных по конверсии в разрезе дней немного (всего 28 дней в анализе);
* значений конверсий распределены ненормально;
* мы не знаем как данные о конверсии в целевое действие по дням распределены в целом.

Сформулируем гипотезы:
*Н0*: конверсия в просмотр контактов у пользователей, установивших приложение по ссылке из yandex, и пользователей, установивших приложение по ссылке из google, равна.     
*Н1*: конверсия в просмотр контактов у пользователей, установивших приложение по ссылке из yandex, и пользователей, установивших приложение по ссылке из google, отличается.    

```python
alpha = 0.05  # критический уровень статистической значимости

results = st.mannwhitneyu(google, yandex, alternative="two-sided")[1]

print('p-значение: ', "{0:.3f}".format(results))

if results < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print(
        'Не получилось отвергнуть нулевую гипотезу'
    ) 
    
print('Относительная разница в конверсии в целевое действие у юзеров из yandex по отношению к google: ', 
      "{0:.1%}".format(yandex.mean()/google.mean()-1))
```

#### Выводы

* Поскольку p-value больше критического уровня значимости отвергнуть нулевую гипотезу не получилось. Статистически значимые отличия коверсии в целевое действие между юзерами, пришедшими из yandex и google, не установлены.      
* В относительном выражении конверсия у юзеров из yandex в среднем больше, чем у юзеров из google на 3,6%.

### Статистическая значимость различий в конверсии в целевое действие между юзерами, проводящими много и мало времени в приложении

Также узнаем ли статистическая значимость в конверсии в целевое действие между юзерами, проводящими много и мало времени в приложении.

В переменной `cr_hyp_time` сохраним датафрейм с данными о количестве юзеров, впервые совершивших в каждый из анализируемых дней целевое действие, в разрезе кластеров по длительности пребывания в приложении:

```python
cr_hyp_time = profiles.pivot_table(
    index='time_cluster',
    columns='dt',
    values='first_ta',
    aggfunc='count'
)

cr_hyp_time
```

В переменной `sizes` сохраним данные о количестве юзеров из каждого кластера на каждый из дней:

```python
sizes = (
    profiles
    .groupby(['time_cluster', 'dt'])['user_id']
    .count()
    .to_frame()
    .reset_index()
)

sizes = (
    sizes
    .pivot_table(
        index='dt', 
        columns='time_cluster', 
        values='user_id')
    .reset_index()
    .set_index('dt')
    .T)

sizes
```

Рассчитаем конверсию на каждый из дней:

```python
cr_hyp_time = cr_hyp_time.div(sizes).T.reset_index()
cr_hyp_time.head()
```

Оценим распределения получившихся значений конверсий:

```python
fig, ax = plt.subplots()
ax.hist(cr_hyp_time[1], alpha=0.5, label='5min/session')
ax.hist(cr_hyp_time[0], alpha=0.5, label='20min/session')
ax.set(title='Распределение значений конверсий по кластерам времени в приложении')
plt.legend();
```

Данные о показателях конверсии имеют ненормальное распределение.    
Сохраним в переменных `cluster_0` (меньше времени в приложении) и `cluster_1` (больше времени в приложении) данные для каждого кластера:

```python
cluster_0 = cr_hyp_time[0]
cluster_1 = cr_hyp_time[1]
```

Используем для проверки гипотез также тест Манна-Уитни.

*Н0*: конверсия в просмотр контактов у пользователей, проводящих больше времени в приложении, и у пользователей, проводящих меньше времени в приложении, равна.      
*Н1*: конверсия в просмотр контактов у пользователей, проводящих больше времени в приложении, и у пользователей, проводящих меньше времени в приложении, отличается.    

```python
alpha = 0.05  # критический уровень статистической значимости

results = st.mannwhitneyu(cluster_0, cluster_1, alternative="two-sided")[1]

print('p-значение: ', "{0:.3f}".format(results))

if results < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print(
        'Не получилось отвергнуть нулевую гипотезу'
    ) 

print('Относительная разница в конверсии в целевое действие у кластера 1 по отношению к кластеру 0: ', 
      "{0:.1%}".format(cluster_0.mean()/cluster_1.mean()-1))
```

#### Выводы

* Поскольку p-value больше критического уровня значимости отвергнуть нулевую гипотезу не получилось. Статистически значимые отличия коверсии в целевое действие между кластерами по времени, проведённом в приложении, не установлены.      
* В относительном выражении конверсия у юзеров из кластера больше проводящих времени в приложении в среднем выше, чем у юзеров из кластера меньше проводящих времени в приложении на 4,5%.
